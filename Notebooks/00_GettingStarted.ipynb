# Getting started with multitaperpy/Multitaper.jl

Here is how you would simply compute the multitaper power spectrum with default values of the parameters. 

```python
from julia.api import Julia
jl = Julia(compiled_modules=False)

from multitaperpy import multispec
import numpy as np

x = np.random.rand(1024)

S = multispec.multispec(x)
```

```python
import matplotlib.pyplot as plt

plt.semilogy(S.f, S.S)
```
